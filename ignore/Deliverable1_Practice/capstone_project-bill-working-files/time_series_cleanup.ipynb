```python
#import standard libraries

import pandas as pd
import matplotlib as plt
import datetime as dt
```

```python
# read data files

url = "time_seris_deaths_state.csv"
covid_death_df = pd.read_csv(url)
```

```python
# verify the read and look at data....

covid_death_df.head()
```

```python
#verify data types in the data frame

covid_death_df.dtypes
```

```python
#drop columns not needed for state level summary and possible machine learning model

covid_death_df.drop(columns=['UID','iso2','iso3','code3','FIPS','Country_Region','Lat','Long_','Combined_Key'], axis=1, inplace=True)
```

```python
#check the resulting data frame

covid_death_df.head()
```

```python
# group and sum by date

covid_state_summary_df = covid_death_df.groupby('Province_State').sum()
```

```python
#verify resulting datafrome

covid_state_summary_df.head()
```

```python
# write the summary dataframe to a csv file for use - keep(state) the index, keep headers

covid_state_summary_df.to_csv("covid_state_summary.csv", index=True, header=True, sep =',')
```

```python
# transpose the dataframe to have states as column headings and covid data by state as rows

covid_summary_TP_df = covid_state_summary_df.transpose(copy=False)
```

```python
covid_summary_TP_df.head()
```

```python
# write the transposed summary dataframe to csv - keep headers, index

covid_summary_TP_df.to_csv("covid_state_summary_TP.csv", index=True, header=True, sep=',')
```
