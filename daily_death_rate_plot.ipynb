```python
import pandas as pd
import matplotlib as plt
import datetime as dt
import numpy as np
```

```python
url = "covid_death_daily_rate_by_state_TP.csv"
covid_death_df = pd.read_csv(url)
```

```python
covid_death_df.head()
```

```python
covid_death_df = covid_death_df.drop(labels=0)
```

```python
covid_death_df.plot(kind='line', figsize=(12,8) ,x='Unnamed: 0',legend=False, title='State Covid Daily Deaths', xlabel='Date', ylabel='Covid Deaths by Day')
```

```python
url = "Covid_summary_stats_by_state.csv"
covid_stats_df = pd.read_csv(url)
```

```python
covid_stats_df.head()
```

```python
covid_stats_df['case_per_1k'] = (covid_stats_df['total_cases']*1000)/covid_stats_df['population']
covid_stats_df['death_per_1k'] = (covid_stats_df['total_deaths']*1000)/covid_stats_df['population']
```

```python
covid_stats_df.head()
```

```python
stats_df = covid_stats_df.describe()
```

```python
stats_df.head()
```

```python
stats_df.plot(kind='box',y=['populaton','total_cases','total_deaths'], logy=True )
```

```python
covid_stats_df.plot(kind='line', figsize=(17,8) ,x='State',y=['case_rate','death_rate'],legend=False, 
                    title='State Case/Death Rate', xlabel='State', ylabel='Covid Confirmed Cases/Deaths',
                    rot = '90', fontsize='8',subplots=True)
```

```python
covid_stats_df.plot(kind='box', figsize=(12,8) ,y=['case_rate','death_rate', 'deaths_per_case'],legend=False, 
                    title='State Cases/Deaths Rate ', xlabel='State',
                    sort_columns = True, logy = True)
```

```python
covid_stats_df.plot(kind='box', figsize=(12,8) ,y=['total_cases','total_deaths'],legend=False, 
                    title='State Total Cases/Deaths', xlabel='State',
                    sort_columns = True, logy = True)
```

```python
covid_stats_df.plot(kind='hist', figsize=(12,8) ,y=['case_rate','death_rate'],legend=False, 
                    title='Distribution Death Rate vs Case Rate')
```

```python
corr = covid_stats_df.corr()
```
