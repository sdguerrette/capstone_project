```python
import pandas as pd
```

```python
pol_df = pd.read_csv('../modified_data/dem_vote_pct.csv', dtype = {'county_fips': 'str'})
pol_df = pol_df.set_index('county_fips')
pol_df
```

```python
del_1_data = pd.read_csv('../modified_data/merged_cleaned.csv', dtype = {'FIPS': 'int'})
del_1_data['FIPS'] = del_1_data['FIPS'].astype('str')
del_1_data['FIPS'] = del_1_data['FIPS'].str.zfill(5)
del_1_data = del_1_data.set_index('FIPS')
```

```python
merged_df = del_1_data.join(pol_df, how='left')
merged_df
```

```python
missing_df = merged_df.loc[merged_df.isnull().any(axis=1)]
missing_df
```

```python
new_merged_df = merged_df.dropna(how='any')
new_merged_df
```

```python
new_merged_df.to_csv('../modified_data/pol_svi_sc_merged.csv')
```
