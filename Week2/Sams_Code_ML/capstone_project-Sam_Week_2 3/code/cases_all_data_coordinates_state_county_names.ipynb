```python
import pandas as pd
```

```python
cases = pd.read_csv("../merged_data/cases_all_data.csv")
cases
```

```python
#add back lat, long, state name, county name 
df = pd.read_csv("../data_sources/time_series_cases_state.csv")
location = df[["FIPS", "Combined_Key", "Province_State", "Lat", "Long_"]]
location.head(5)
```

```python
#merge with death df
cases_all_data_location = pd.merge(cases, location, on="FIPS", how="outer")
cases_all_data_location
```

```python
cases_all_data_location.dropna()
```

```python
county_fips= pd.read_csv("../fips-codes-master/county_fips_master.csv", encoding= "ISO-8859-1")
county_fips
```

```python
county= county_fips[["FIPS", "county_name"]]
county
```

```python
cases_all_data_location = pd.merge(cases_all_data_location, county, on="FIPS", how="outer" )
cases_all_data_location 
```

```python
cases_all_data_location.dropna()
```

```python
cases_all_data_location.to_csv("../merged_data/cases_all_data_location.csv")
```
