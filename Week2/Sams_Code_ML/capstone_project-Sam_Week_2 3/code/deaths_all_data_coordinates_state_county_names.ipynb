```python
import pandas as pd
```

```python
deaths = pd.read_csv("../merged_data/deaths_all_data.csv")
deaths
```

```python
#add back lat, long, state name, county name 
df = pd.read_csv("../data_sources/time_series_cases_state.csv")
location = df[["FIPS", "Combined_Key", "Province_State","Admin2", "Lat", "Long_"]]
location.head(5)
```

```python
#merge with death df
deaths_all_data_location = pd.merge(deaths, location, on="FIPS", how="left")
deaths_all_data_location
```
