```python
#import dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
#read in merged and cleaned data
df = pd.read_csv('../modified_data/pol_svi_sc_merged.csv')
df
```

```python
deaths_df = pd.read_csv('../modified_data/death_data.csv')
deaths_df = deaths_df[['FIPS','3/31/21']]
deaths_df = deaths_df.set_index('FIPS')
deaths_df
```

```python
#set index to FIPS
df = df.set_index(df['FIPS'])
df= df.drop('FIPS', axis=1)
```

```python
#rename target column
df = df.drop('3/31/21', axis=1)
df
```

```python
df = df.merge(deaths_df, how='left', left_index=True, right_index=True)
df
```

```python
df['first_year_deaths'] = df['3/31/21'].rename()
```

```python
#create variable for case% for each counties population
df['death_pct'] = df['first_year_deaths']/df['E_TOTPOP']*100
df['death_pct'].head()
```

```python
df['death_pct'].describe()
```

```python
# bin and cut the case_pct column into 2 classifications
# q = df['case_pct'].quantile(.75)
q = df['death_pct'].quantile(.9)
bins = [0, q , 1]
labels = ['low','high']
df['death_class'] = pd.cut(df['death_pct'], bins, labels = labels)
df['death_class']
```

```python
df['death_class'].value_counts()
```

```python
#drop unneeded columns
df = df.drop('death_pct', axis =1)
df
```

```python
#turn case % classifications into binary 
df = pd.get_dummies(df, columns = ['death_class'])
df
```

```python
impact_cols = ['Empathy',
 'EPL_GROUPQ',
 'Agreeableness',
 'Employment Rate',
 'EPL_UNEMP',
 'Openness',
 'E_AGE65',
 'EP_AGE17',
 'Conscientiousness',
 'SPL_THEME4',
 'Income Per Capita',
 'E_POV',
 'first_year_deaths',
 'E_MINRTY',
 'Conflict Awareness',
 'Extraversion',
 'EP_SNGPNT',
 'F_MOBILE',
 'Work Ethic',
 'E_AGE17',
 'EP_PCI',
 'Risk Taking',
 'E_UNEMP',
 'AREA_SQMI',
 'E_NOHSDP',
 'EPL_PCI',
 'E_HU',
 'E_NOVEH',
 'E_LIMENG',
 'EPL_SNGPNT',
 'E_UNINSUR',
 'EPL_AGE17',
 'Belief In Science',
 'EPL_LIMENG',
 'EPL_NOVEH',
 'Entrepreneurship',
 'SPL_THEME3',
 'E_HH',
 'EP_MUNIT',
 'EPL_MOBILE',
 'Collectivism',
 'RPL_THEME2',
 'E_MUNIT',
 'EPL_POV',
 'EP_UNINSUR',
 'dem_pct',
 'F_SNGPNT',
 'EPL_MUNIT',
 'EP_NOHSDP',
 'EPL_DISABL',
 'Hopefulness',
 'E_MOBILE']
```

```python
#seperate targets and features
X = df[impact_cols].values
y=df['death_class_high'].values
```

```python
#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
# hidden_nodes_layer1 = 100
# hidden_nodes_layer2 = 50
hidden_nodes_layer1 = 150
hidden_nodes_layer2 = 100
# hidden_nodes_layer3 = 50

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))
# nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))
# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
y_pred=nn.predict(X_test)
con_mat = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)
con_mat
```
