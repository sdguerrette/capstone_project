```python
import pandas as pd
```

```python
hospitals_df = pd.read_csv('../../data/data_sources/Hospitals.csv')
hospitals_df
```

```python
hospitals_df['NAICS_DESC'].value_counts()
```

```python
naics_keep = ['GENERAL MEDICAL AND SURGICAL HOSPITALS',
             "CHILDREN'S HOSPITALS, GENERAL"]
hospitals_df = hospitals_df.loc[(hospitals_df['NAICS_DESC']=='GENERAL MEDICAL AND SURGICAL HOSPITALS')|(hospitals_df['NAICS_DESC']=="CHILDREN'S HOSPITALS, GENERAL")]
hospitals_df
```

```python
hospitals_df.columns.to_list()
```

```python
missing = hospitals_df.loc[hospitals_df['BEDS']==-999]
hospital_beds = hospitals_df.drop(missing.index, axis=0)
hospital_beds
```

```python
hospital_beds = hospital_beds[['COUNTYFIPS','BEDS']]
hospital_beds
```

```python
hospital_beds['COUNTYFIPS'].nunique()
```

```python
county_beds = hospital_beds.groupby(['COUNTYFIPS']).sum()
county_beds
```

```python
county_beds = county_beds.drop('NOT AVAILABLE', axis=0)
```

```python
county_beds.to_csv('../../data/cleaned_data/FIPS_hospital_beds.csv')
```
