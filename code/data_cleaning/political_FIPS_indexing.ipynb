```python
import pandas as pd
```

```python
pol_df = pd.read_csv('../../data/data_sources/election_results_county.csv', dtype = {'county_fips':'str'})
pol_df.head()
```

```python
# pol_df = pol_df.drop(columns = ['version','mode','state','candidate'])
```

```python
latest_pol_df = pol_df.loc[pol_df['year']==2020]
latest_pol_df
```

```python
dem_df = latest_pol_df.loc[latest_pol_df['party']=='DEMOCRAT']
dem_df.loc[dem_df['mode']!= 'TOTAL']
```

```python
county_dem_group = dem_df.groupby(by =['county_fips']).sum()
county_dem_group
```

```python
county_dem_group['dem_pct'] = county_dem_group['candidatevotes']/county_dem_group['totalvotes']*100
county_dem_group
```

```python
dem_vote_pct = county_dem_group['dem_pct']
dem_vote_pct
```

```python
dem_vote_pct.to_csv('../../data/cleaned_data/dem_vote_pct.csv')
```
