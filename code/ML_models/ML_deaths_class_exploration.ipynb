```python
#import dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
#read in merged and cleaned data
df = pd.read_csv('../../data/merged_cleaned_data/deaths_merged_data.csv')
df
```

```python
#set index to FIPS
df = df.set_index(df['FIPS'])
df= df.drop(columns = ['FIPS','Unnamed: 0'])
```

```python
# examine dataframe
df
```

```python
#create variable for case% for each counties population
df['death_pct'] = df['first_year_deaths']/df['E_TOTPOP']*100
df['death_pct'].head()
```

```python
df['death_pct'].describe()
```

```python
# bin and cut the case_pct column into 2 classifications
q = df['death_pct'].quantile(.9)
bins = [0, q , 1]
labels = ['low','high']
df['death_class'] = pd.cut(df['death_pct'], bins, labels = labels)
df['death_class']
```

```python
df['death_class'].value_counts()
```

```python
#drop unneeded columns
df = df.drop('death_pct', axis =1)
df
```

```python
#turn case % classifications into binary 
df = pd.get_dummies(df, columns = ['death_class'])
df
```

```python
#seperate targets and features
# X = df[impact_cols].values
X = df.drop(columns = ['death_class_high','death_class_low','first_year_deaths']).values
y=df['death_class_high'].values
```

```python
#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 100
hidden_nodes_layer2 = 80
# hidden_nodes_layer3 = 50

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))
# nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))
# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=150)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
y_pred=nn.predict(X_test_scaled)
con_mat = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)
con_mat
```
