```python
import pandas as pd
```

```python
deaths_df = pd.read_csv('../modified_data/death_data.csv')
pol_df = pd.read_csv('../modified_data/dem_vote_pct.csv')
svi_df =pd.read_csv('../modified_data/svi_cleaned')
sc_df = pd.read_csv('../modified_data/social_context_FIPS.csv')
beds_df = pd.read_csv('../modified_data/FIPS_hospital_beds.csv')
```

## SVI Data

```python
svi_df['FIPS'] = svi_df['FIPS'].astype('str')
svi_df['FIPS'] = svi_df['FIPS'].str.zfill(5)
svi_df = svi_df.drop('LOCATION', axis=1)
```

## SC Data

```python
sc_df
```

```python
sc_df['FIPS'] = sc_df['IBRC_Geo_ID'].rename()
sc_df['FIPS'] = sc_df['FIPS'].astype('str')
sc_df['FIPS'] = sc_df['FIPS'].str.zfill(5)
sc_df = sc_df.drop(['IBRC_Geo_ID','Income Mobility'], axis =1)
sc_df
```

## Pol Data

```python
pol_df
```

```python
pol_df = pol_df.rename(columns={'county_fips':'FIPS'})
pol_df['FIPS'] = pol_df['FIPS'].astype('str').str.zfill(5)
pol_df
```

## Beds Data

```python
beds_df.dtypes
```

```python
beds_df = beds_df.rename(columns={'COUNTYFIPS':'FIPS'})
beds_df['FIPS'] = beds_df['FIPS'].astype('str').str.zfill(5)
beds_df
```

## Deaths Data


```python
deaths_df.dtypes
```

```python
deaths_df = deaths_df.drop(columns = ['Admin2','Province_State','Unnamed: 0'])
deaths_df = deaths_df.rename(columns={'3/31/21':'first_year_deaths'})
deaths_df
```

```python
deaths_df = deaths_df.dropna(how='any')
deaths_df['FIPS'] = deaths_df['FIPS'].astype('int')
deaths_df['FIPS'] = deaths_df['FIPS'].astype('str').str.zfill(5)
deaths_df
```

## Merge all dataframes on FIPS


```python
svi_sc = svi_df.merge(sc_df, how='left', on = 'FIPS')
svi_sc.isnull().sum()
```

```python
svi_sc = svi_sc.dropna(how='any')
svi_sc
```

```python
svi_sc_pol = svi_sc.merge(pol_df,how='left', on = 'FIPS')
svi_sc_pol = svi_sc_pol.dropna(how='any')
svi_sc_pol
```

```python
svi_sc_pol_beds = svi_sc_pol.merge(beds_df, how='left', on = 'FIPS')
svi_sc_pol_beds.isnull().sum()
```

```python
svi_sc_pol_beds = svi_sc_pol_beds.fillna(value=0)
svi_sc_pol_beds
```

```python
all_merged = svi_sc_pol_beds.merge(deaths_df, how='left', on='FIPS')
all_merged.isnull().sum()
```

```python
all_merged
```
