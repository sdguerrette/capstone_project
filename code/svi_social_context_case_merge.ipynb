```python
#import dependencies
import pandas as pd
```

```python
#read in svi/socialcontext and case rate data
case_df = pd.read_csv('../modified_data/case_data.csv')
svi_cs_df = pd.read_csv('../modified_data/svi_social_context_merged')
```

```python
#drio unneeded columns and set index for FIPS
case_df = case_df.drop('Unnamed: 0', axis=1)
case_df = case_df.set_index(case_df['FIPS'])
case_df = case_df.drop('FIPS', axis=1)
case_df
```

```python
#Set index for FIPS
svi_cs_df = svi_cs_df.set_index(svi_cs_df['FIPS'])
svi_cs_df = svi_cs_df.drop(columns=['FIPS'])
svi_cs_df
```

```python
#merge files based on FIPS index
merged_df = case_df.merge(svi_cs_df, how='left', left_index=True, right_index=True)
merged_df
```

```python
#check for nulls
merged_df.isnull().sum().to_list()
```

```python
#drop column because of the large number of null values, drop rows with null values
cleaned_df = merged_df.drop('Income Mobility', axis=1)
cleaned_df = cleaned_df.dropna()
cleaned_df.columns.to_list()
```

```python
#drop unneeded columns
cleaned_df = cleaned_df.drop(columns=['Admin2','Province_State','LOCATION'])
```

```python
cleaned_df.dtypes.to_list()
```

```python
#write to CSV
cleaned_df.to_csv('../modified_data/merged_cleaned.csv')
```
