```python
#import dependencies
import pandas as pd
```

```python
#read in svi and social context data
svi_df = pd.read_csv('../data_sources/SVI2018_US_COUNTY.csv')
sc_df = pd.read_csv('../modified_data/social_context_FIPS.csv')
```

```python
#set index to FIPS
svi_df = svi_df.set_index('FIPS')
svi_df.head()
```

```python
#drop unneeded columns
svi_df = svi_df.drop(columns = ['ST','STATE','ST_ABBR','COUNTY'])
```

```python
#drop row with error - per data source documentation
svi_df = svi_df.drop(index = 35039)
```

```python
svi_df.head()
```

```python
#prepare to drop the margin of error columns
cols = svi_df.drop(columns = ['LOCATION','AREA_SQMI'])
```

```python
cols
```

```python
# select all column names that start with an M - these are margin of error columns
col_list = cols.columns.to_list()
col_series = pd.Series(col_list)
drop_str = r'^M+.'
drop_form = col_series.str.contains(drop_str)
drop_col = col_series[drop_form].to_list()

    
```

```python
#drop the margin of error columns
svi_df = svi_df.drop(columns = drop_col)
svi_df
```

```python
sc_df.head()
```

```python
#set index to match FIPS
sc_df = sc_df.set_index('IBRC_Geo_ID')
sc_df.head()
```

```python
#merge dataframes by index
merged_df = svi_df.merge(sc_df, how = 'left', left_index=True, right_index=True, sort=True)
```

```python
merged_df.columns.to_list()
```

```python
#check for null values
missing_df = merged_df[merged_df.isnull().any(axis=1)]
```

```python
merged_df.to_csv('../modified_data/svi_social_context_merged')
```
