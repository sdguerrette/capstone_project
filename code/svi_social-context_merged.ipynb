```python
import pandas as pd
```

```python
svi_df = pd.read_csv('../data_sources/SVI2018_US_COUNTY.csv')
sc_df = pd.read_csv('../modified_data/social_context_FIPS.csv')
```

```python
svi_df = svi_df.set_index('FIPS')
svi_df.head()
```

```python
svi_df = svi_df.drop(columns = ['ST','STATE','ST_ABBR','COUNTY'])
```

```python
svi_df = svi_df.drop(index = 35039)
```

```python
svi_df.head()
```

```python
cols = svi_df.drop(columns = ['LOCATION','AREA_SQMI'])
```

```python
cols
```

```python
col_list = cols.columns.to_list()
col_series = pd.Series(col_list)
drop_str = r'^M+.'
drop_form = col_series.str.contains(drop_str)
drop_col = col_series[drop_form].to_list()

    
```

```python
svi_df = svi_df.drop(columns = drop_col)
svi_df
```

```python
merged_df = svi_df.merge(sc_df, how = 'left', left_index=True, right_index=True, sort=True)
```
