```python
import pandas as pd
```

```python
df = pd.read_csv('../data_sources/co-est2021-alldata.csv', encoding = 'latin-1')
df.head()
```

```python
state_rows = df.loc[df['COUNTY']==0]
```

```python
state_rows
```

```python
counties_df = df.drop(index = state_rows.index)
```

```python
counties_df['COUNTY'] = counties_df['COUNTY'].astype('str')
```

```python
counties_df['COUNTY'] = counties_df['COUNTY'].str.zfill(3)
counties_df.head()
```

```python
counties_df['STATE'] = counties_df['STATE'].astype('str')
```

```python
counties_df['FIPS'] = counties_df['STATE']+counties_df['COUNTY']
```

```python
counties_df = counties_df.set_index(counties_df['FIPS'])
```

```python
counties_df.loc['22113']
```

```python
counties_df = counties_df.drop(columns = ['SUMLEV','REGION','DIVISION','STATE','COUNTY','FIPS','STNAME','CTYNAME'])
```

```python
counties_df.count()[1]
```

```python
counties_df.to_csv('../modified_data/co-est2021-alldata_FIPS.csv')
```
